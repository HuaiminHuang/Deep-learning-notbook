# 绘制折线图

```python
#用于加载中文字体
from matplotlib.pyplot import rcParams 
rcParams['font.family'] = 'SimHei'

# 设定数据
x_ax = ['1月', '2月', '3月', ]
y_ax = [110, 230, 185]

# 配置参数
plt.figure(figsize = (4,3))
plt.title('个人销售记录-1表', fontsize = 10)
plt.xlabel('月份', fontsize = 8)
plt.ylabel('销售额', fontsize = 8)
plt.grid(True, linestyle="--", alpha=0.8) #网格线

#画图展示
plt.plot(x_ax, y_ax, label='产品A')
plt.legend(loc = 'upper right')
plt.show()
```

# 绘制散点图

```python
import matplotlib.pyplot as plt
import random

plt.rcParams['font.sans-serif'] = ['SimHei']      # 黑体
plt.figure(figsize=(4,3))

#设定数据(离散随机-均匀分布加高斯噪声)
x = []
y = []
for k in range(100):
    temp = random.uniform(5,18)
    x.append(temp)
    y.append(2*temp + random.gauss(0, 3))

plt.scatter(x, y,
        label = "离散样本A")
plt.legend(loc = "lower right", fontsize = 12)
plt.show()
```

# 箱庭图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 中文与风格
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 生成成绩
mean  = np.array([75, 70, 78])
cov   = np.array([[64, 35, 30],
                  [35,144, 40],
                  [30, 40,100]])
scores = np.clip(np.random.multivariate_normal(mean, cov, size=20),
                 0, 100).round().astype(int)

df = pd.DataFrame(scores,
                  columns=['语文', '数学', '英语'],
                  index=[f'S{i:02d}' for i in range(1, 21)])

# 画图
plt.figure(figsize=(4, 3))
plt.boxplot(df.values, labels=df.columns) ##老版本不支持tick_labels="" 改———>labels="" 
plt.title('各科成绩箱线图')
plt.ylabel('分数')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
