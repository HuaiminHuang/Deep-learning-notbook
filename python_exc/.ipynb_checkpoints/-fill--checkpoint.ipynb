```python
import pandas as pd
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np

# 加载数据集
url = "https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
recovered_df = pd.read_csv(url)

# 检查数据结构
print(f"数据集维度: {recovered_df.shape}")
print(f"日期列范围: {recovered_df.columns[4]} 至 {recovered_df.columns[-1]}")
```

```python
# 计算每日全球康复总数
global_recovered = recovered_df.iloc[:, 4:].sum(axis=0)
global_recovered.index = pd.to_datetime(global_recovered.index)  # 转换日期索引
global_recovered = global_recovered.sort_index()  # 确保日期排序

# 聚焦2021年8月
august_2021 = global_recovered['2021-08']
print("2021年8月康复数据统计:")
print(august_2021.describe())

# 检查是否出现0值
zero_days = august_2021[august_2021 == 0]
if not zero_days.empty:
    print(f"\n警告: 在以下日期康复数据为0:")
    print(zero_days)
else:
    print("\n2021年8月没有康复数据为0的日期")
```

```python
plt.figure(figsize=(14, 8))
plt.plot(global_recovered.index, global_recovered.values, label='Global Recovered')

# 突出显示2021年8月
august_mask = (global_recovered.index >= '2021-08-01') & (global_recovered.index <= '2021-08-31')
plt.fill_between(global_recovered.index[august_mask], 
                 global_recovered[august_mask].min(), 
                 global_recovered[august_mask].max(), 
                 color='yellow', alpha=0.3, label='August 2021 Focus')

# 标记异常点（如果存在）
if not zero_days.empty:
    plt.scatter(zero_days.index, zero_days.values, 
                color='red', s=100, zorder=5, label='Zero Values')

plt.title('Global COVID-19 Recovered Cases (JHU Dataset)')
plt.xlabel('Date')
plt.ylabel('Recovered Cases')
plt.legend()
plt.grid(True)
plt.yscale('log')  # 使用对数刻度更易观察变化
plt.show()
```

```python
# 检查特定国家/地区是否导致问题
def check_country_recovery(country):
    country_data = recovered_df[recovered_df['Country/Region'] == country]
    country_ts = country_data.iloc[:, 4:].sum(axis=0)
    country_ts.index = pd.to_datetime(country_ts.index)
    return country_ts['2021-08']

# 检查主要国家
major_countries = ['US', 'India', 'Brazil', 'France', 'UK', 'Russia', 'Germany']
for country in major_countries:
    aug_data = check_country_recovery(country)
    print(f"\n{country} 2021年8月康复数据:")
    print(f"最小值: {aug_data.min():,}, 最大值: {aug_data.max():,}")
    zero_days = aug_data[aug_data == 0]
    if not zero_days.empty:
        print(f"  异常零值日期: {zero_days.index.strftime('%Y-%m-%d').tolist()}")
```

```python
# 加载康复数据
url = "https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
recovered_df = pd.read_csv(url)

# 计算全球每日康复总数
global_recovered = recovered_df.iloc[:, 4:].sum(axis=0)
global_recovered.index = pd.to_datetime(global_recovered.index)

# 转换为Prophet所需格式
prophet_df = pd.DataFrame({
    'ds': global_recovered.index,
    'y': global_recovered.values
})

# 过滤到2021年夏季前的数据 (截至2021-06-30)
train_df = prophet_df[prophet_df['ds'] <= '2021-06-30']

print(f"训练数据量: {len(train_df)} 天")
print(f"最后训练日期: {train_df['ds'].max().date()}")
```

```python
# 创建并训练Prophet模型
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    changepoint_prior_scale=0.5,
    seasonality_mode='multiplicative'
)

model.fit(train_df)

# 创建预测框架 (预测到2021-12-31)
future = model.make_future_dataframe(periods=184)  # 2021下半年共184天
forecast = model.predict(future)

# 提取预测结果
prophet_forecast = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
prophet_forecast = prophet_forecast.set_index('ds')
```

```python
# 准备ARIMA数据
arima_data = train_df.set_index('ds')['y']

# 寻找最佳(p,d,q)参数 - 使用网格搜索
best_aic = np.inf
best_order = None

for p in range(0, 3):
    for d in range(0, 2):
        for q in range(0, 3):
            try:
                model = ARIMA(arima_data, order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic:
                    best_aic = results.aic
                    best_order = (p, d, q)
            except:
                continue

print(f"最佳ARIMA参数: {best_order}, AIC: {best_aic:.2f}")

# 使用最佳参数训练模型
best_model = ARIMA(arima_data, order=best_order)
best_results = best_model.fit()

# 预测到2021年底
arima_forecast = best_results.get_forecast(steps=184)
arima_ci = arima_forecast.conf_int()
```

```python
plt.figure(figsize=(14, 8))

# 绘制历史数据
plt.plot(global_recovered.index, global_recovered.values, 
         'b-', label='Actual Recovered Cases', alpha=0.7)

# 绘制Prophet预测
plt.plot(prophet_forecast.index, prophet_forecast['yhat'], 
         'r--', label='Prophet Forecast')
plt.fill_between(prophet_forecast.index, 
                 prophet_forecast['yhat_lower'], 
                 prophet_forecast['yhat_upper'], 
                 color='red', alpha=0.1)

# 绘制ARIMA预测
plt.plot(arima_ci.index, arima_forecast.predicted_mean, 
         'g-.', label='ARIMA Forecast')
plt.fill_between(arima_ci.index, 
                 arima_ci.iloc[:, 0], arima_ci.iloc[:, 1], 
                 color='green', alpha=0.1)

# 标记关键时间点
plt.axvline(pd.to_datetime('2021-06-30'), color='gray', linestyle='--', 
            label='Prediction Start')
plt.axvline(pd.to_datetime('2021-12-31'), color='black', linestyle='--', 
            label='Prediction End')

# 添加Delta变种注释
plt.annotate('Delta Variant Dominance', 
             xy=(pd.to_datetime('2021-07-15'),1.6e8), 
             xytext=(pd.to_datetime('2021-05-01'), 1.5e8),
             arrowprops=dict(arrowstyle='->', color='black'))

plt.title('Global COVID-19 Recovered Cases Forecast (Summer to End of 2021)')
plt.xlabel('Date')
plt.ylabel('Recovered Cases')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.savefig('Recovered Cases Forecast.png')
plt.show()
plt.close()

# 输出预测摘要
print("\n预测摘要 (2021-12-31):")
print(f"Prophet预测: {prophet_forecast.loc['2021-12-31', 'yhat']:,.0f} 例")
print(f"ARIMA预测: {arima_forecast.predicted_mean[-1]:,.0f} 例")
print(f"实际值 (2021-12-31): {global_recovered.loc['2021-12-31']:,.0f} 例")
```
